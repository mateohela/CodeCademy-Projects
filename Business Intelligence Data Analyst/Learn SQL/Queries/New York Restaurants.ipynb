```python
-- 11 Use a CASE statement to change the rating system to:
--   review > 4.5 is Extraordinary
--   review > 4 is Excellent
--   review > 3 is Good
--   review > 2 is Fair
-- Everything else is Poor

SELECT name, cuisine, price, CASE
  WHEN review > 4.5 THEN 'Extraordinary'
  WHEN review > 4 THEN 'Excellent'
  WHEN review > 3 THEN 'Good'
  WHEN review > 2 THEN 'Fair'
  ELSE 'Poor'
  END AS 'Named reviews'
FROM nomnom
LIMIT 10;

-- Results
Query Results
name	cuisine	price	Named reviews
Peter Luger Steak House	Steak	$$$$	Excellent
Jongro BBQ	Korean	$$	Excellent
Pocha 32	Korean	$$	Good
Nom Wah Tea Parlor	Chinese	$	Excellent
Roberta's	Pizza	$$	Excellent
Speedy Romeo	Pizza	$$	Excellent
Bunna Cafe	Ethiopian	$$	Extraordinary
Massawa	Ethiopian	$$	Good
Buddha Bodai	Vegetarian	$$	Excellent
Nan Xiang Xiao Long Bao	Chinese	$	Excellent
```
